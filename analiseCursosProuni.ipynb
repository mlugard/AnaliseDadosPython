```python
import pandas as pd

cursos_df = pd.read_csv("cursos-prouni.csv")
```

```python
display(cursos_df)
```

```python
#Exibir os primeiros 15 resultados da tabela
display(cursos_df.head(15))
```

```python
#Filtrar por apenas uma coluna
display(cursos_df['curso_busca'])
```

```python
eng_df = cursos_df.loc[cursos_df['curso_busca'] == 'Engenharia Elétrica'].head(10)
display(eng_df)
```

```python
#Filtrar os cursos de Engenharia elétrica com nota de corte inferior a 700 pontos na cidade de São Paulo
eng_df = cursos_df[['curso_busca', 'cidade_busca', 'nota_integral_ampla']].loc[(cursos_df['curso_busca'] == 'Engenharia Elétrica') & (cursos_df['cidade_busca'] == 'Sao Paulo') & (cursos_df['nota_integral_ampla'] < 700.00)]
display(eng_df.sort_values(by='nota_integral_ampla'))
```

```python
#Apresentar a maior nota de corte a partir do filtro anterior
eng_df = cursos_df[['curso_busca', 'cidade_busca', 'nota_integral_ampla']].loc[(cursos_df['curso_busca'] == 'Engenharia Elétrica') & (cursos_df['cidade_busca'] == 'Sao Paulo') & (cursos_df['nota_integral_ampla'] < 700.00)]
display(eng_df['nota_integral_ampla'].max())
```

```python
# Como mostrar a média de valor de nota POR CURSO do Brasil todo?
media_por_nota_df = cursos_df[['curso_busca', 'nota_integral_ampla']].groupby(by='curso_busca').mean().round(2)
display(media_por_nota_df)
```
